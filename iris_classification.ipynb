```python
inputs=df.drop(["species"] , axis=1)
target=df["species"]
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test , y_train , y_test=train_test_split(inputs , target , test_size=0.25 , random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

log=LogisticRegression(solver="saga",multi_class="ovr" , max_iter=5000)
penalty=["l1" ,"l2" ,"elasticnet"]
l1_ratio=np.linspace(0,1,20)
C=np.logspace(0,10,20)
param_grid={"penalty":penalty ,"l1_ratio":l1_ratio ,"C":C}

grid_model=GridSearchCV(log , param_grid=param_grid )
grid_model.fit(x_train_scaled , y_train)
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix ,classification_report
```

```python
grid_model.best_params_
```

```python
y_pred=grid_model.predict(x_test_scaled)
```

```python
y_pred
```

```python
accuracy_score(y_test , y_pred)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
classification_report(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(grid_model , x_test_scaled , y_test)
```

# ROC curve

```python
from sklearn.metrics import roc_curve , auc

def plot_multiclass_roc(clf , x_test , y_test , n_classes , figsize=(5 ,5)):
    y_score= clf.decision_function(x_test)
    
    #structures
    fpr=dict()
    tpr=dict()
    roc_auc=dict()
    
    #calculate dummies ones
    y_test_dummies=pd.get_dummies(y_test , drop_first=False).values
    for i in range(n_classes):
        fpr[i] , tpr[i] ,_ = roc_curve(y_test_dummies[: , i] , y_score[: ,i])
        roc_auc[i]=auc(fpr[i], tpr[i])
        
    #roc for each class    
    fig , ax=plt.subplots(figsize=figsize)
    ax.plot([0,1],[0,1] ,"k--")
    ax.set_xlim(0.0 ,1.0)
    ax.set_ylim(0.0 ,1.05)
    ax.set_xlabel({"False positive Rate"})
    ax.set_ylabel({"True positive Rate"})
    ax.set_title("ROC curve")
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i] ,label="ROC curve (area=%0.2f)for label %i" % (roc_auc[i] , i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
    
plot_multiclass_roc(grid_model , x_test_scaled , y_test ,n_classes=3 , figsize=(16 ,10))
```
